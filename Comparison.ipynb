```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('train_logs_with_nulls.json', 'r') as f:
    metrics_train_null = json.load(f)
    
with open('val_logs_with_nulls.json', 'r') as f:
    metrics_valid_null = json.load(f)
    
with open('train_logs_without_nulls.json', 'r') as f:
    metrics_train = json.load(f)
    
with open('val_logs_without_nulls.json', 'r') as f:
    metrics_valid = json.load(f)
```

```python
x_axis_null = np.array([x/516 for x in range(len(metrics_valid_null["dice"]))])
x_axis = np.array([x/204 for x in range(len(metrics_valid["dice"]))])
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 6))
ax[0].plot(x_axis, metrics_valid["dice"], label='val', alpha=0.75)
ax[0].plot(x_axis, metrics_train["dice"], label='train', alpha=0.75)
ax[0].set_xlabel('Epoch')
ax[0].set_ylabel('Dice Loss')
ax[0].set_title('Without Null Images')
ax[0].legend();
ax[1].plot(x_axis_null, metrics_valid_null["dice"], label='val', alpha=0.75)
ax[1].plot(x_axis_null, metrics_train_null["dice"], label='train', alpha=0.75)
ax[1].set_xlabel('Epoch')
ax[1].set_ylabel('Dice Loss')
ax[1].set_title('With Null Images')
ax[1].legend();
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 6))
ax[0].plot(x_axis, np.log(metrics_valid["dice"]), label='val', alpha=0.75)
ax[0].plot(x_axis, np.log(metrics_train["dice"]), label='train', alpha=0.75)
ax[0].set_xlabel('Epoch')
ax[0].set_ylabel('Log(Dice Loss)')
ax[0].set_title('Without Null Images')
ax[0].legend();
ax[1].plot(x_axis_null, np.log(metrics_valid_null["dice"]), label='val', alpha=0.75)
ax[1].plot(x_axis_null, np.log(metrics_train_null["dice"]), label='train', alpha=0.75)
ax[1].set_xlabel('Epoch')
ax[1].set_ylabel('Log(Dice Loss)')
ax[1].set_title('With Null Images')
ax[1].legend();
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 6))
ax[0].plot(np.log(1 + x_axis), np.log(metrics_valid["dice"]), label='val', alpha=0.75)
ax[0].plot(np.log(1 + x_axis), np.log(metrics_train["dice"]), label='train', alpha=0.75)
ax[0].set_xlabel('Log(Epoch)')
ax[0].set_ylabel('Log(Dice Loss)')
ax[0].set_title('Without Null Images')
ax[0].legend();
ax[1].plot(np.log(1 + x_axis_null), np.log(metrics_valid_null["dice"]), label='val', alpha=0.75)
ax[1].plot(np.log(1 + x_axis_null), np.log(metrics_train_null["dice"]), label='train', alpha=0.75)
ax[1].set_xlabel('Log(Epoch)')
ax[1].set_ylabel('Log(Dice Loss)')
ax[1].set_title('With Null Images')
ax[1].legend();
```
